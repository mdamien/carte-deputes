# Chargement des données

```python
# fichiers de https://parltrack.org/dumps
import json, collections

all_people = json.load(open('ep_meps.json'))
people_by_name = {people['Name']['full']: people["UserID"] for people in all_people}
people_by_id = {people["UserID"]: people for people in all_people}

scrutins = json.load(open('ep_votes.json'))
print(len(scrutins), 'votes')
print(len(all_people), 'people')
```

```python
for name, id in people_by_name.items():
    print(name, id)
```

```python
def current_group(people):
    for group in people.get('Groups', []):
        if group['end'].startswith('9999'):
            return group['Organization']

all_people = [people for people in all_people if current_group(people)]
people_by_name = {people['Name']['full']: people["UserID"] for people in all_people}
people_by_id = {people["UserID"]: people for people in all_people}
print(len(people))
```

```python
scrutins = [scrutin for scrutin in scrutins if scrutin['ts'] > '2019-07-02']
print(len(scrutins))
```

```python
def position(scrutin, mepid):
    for position in '+', '-', '0':
        for group, votes in scrutin.get('votes', {}).get(position, {}).get('groups', {}).items():
            for vote in votes:
                if vote.get('mepid') == mepid:
                    return position

def similarity(people1, people2):
    if type(people1) == str:
        people1 = people_by_name[people1]
    if type(people2) == str:
        people2 = people_by_name[people2]
    total = 0
    total_common = 0
    for scrutin in scrutins:
        pos1 = position(scrutin, people1)
        pos2 = position(scrutin, people2)
        if not pos1 or not pos2:
            continue
        if pos1 == pos2:
            total_common += 1
        total += 1
    if total == 0:
        return 0, 0
    return total_common / total, total

print(position(scrutins[0], 4514))
print(similarity("Manon AUBRY", "Gilbert COLLARD"))
```

```python
import pandas as pd

AOC = 'Manon AUBRY'
other = 'Nathalie LOISEAU'

data = []
for i, el in enumerate(people_by_name.items()):
    name, id = el
    print(i)
    sim, total = similarity(AOC, id)
    sim_other, total_other = similarity(other, id)
    if total < 20 or total_other < 20:
        continue
    data.append({
        'sim_aubry': sim,
        'sim_loiseau': sim_other,
        'name': name,
        'party': current_group(people_by_id[id]),
    })
df = pd.DataFrame(data)
import plotly.express as px
# fig = px.scatter_3d(df, x=df.similarité_ruffin, y=df.similarité_bergé, z=df.similarité_abad, color=partis, hover_name='nom')
fig = px.scatter(df, x=df.sim_aubry, y=df.sim_loiseau, color=df.party, hover_name=df.name)
fig.show()
fig.write_html("demo.html")
```
